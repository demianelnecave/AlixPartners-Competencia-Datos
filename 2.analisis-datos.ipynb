**ANÁLISIS EXPLORATORIO DE DATOS**

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

catalogo_productos = pd.read_csv("Datos-limpiados/catalogo_productos.csv")
especificaciones_cajas = pd.read_csv("Datos-limpiados/especificaciones_cajas.csv")
operaciones_planta = pd.read_csv("Datos-limpiados/operaciones_planta.csv")
procurement_cajas = pd.read_csv("Datos-limpiados/procurement_cajas.csv")
```

```python
catalogo_productos.head()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(catalogo_productos['peso_neto_caja'].dropna(), bins=10,
         color='green', rwidth=0.9)
plt.title('Distribución del peso neto de la caja por paquete')
plt.xlabel('Peso neto caja (kg)')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
especificaciones_cajas.head()
```

```python
for grosor in [3, 4.5, 5]:
    cantidad = (especificaciones_cajas['caja_grosor_mm'] == grosor).sum()
    print(f"Grosor {grosor} mm: {cantidad} cajas")
```

```python
plt.figure(figsize=(10, 5))
plt.hist(especificaciones_cajas['utilizacion'].dropna(), bins=15,
         color='green', rwidth=0.9)
plt.title('Distribución de utilización de pallet por caja')
plt.xlabel('Porcentaje (%)')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
operaciones_planta.head()
```

```python
cols = [c for c in operaciones_planta.columns if c.startswith('volumen_producto_planta_')]
sums = {}
for c in cols:
    vals = pd.to_numeric(operaciones_planta[c], errors='coerce').fillna(0)
    sums[c.replace('volumen_producto_planta_', '')] = vals.sum()
    
serie = pd.Series(sums).sort_values(ascending=False)

print("Productos enviados por planta (volumen total)")
print("-" * 50)
for planta, total in serie.items():
    print(f"{planta:20s} {total:12,.2f} m3")
print("-" * 50)
```
