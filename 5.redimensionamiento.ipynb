#### **REDIMENSIONAMIENTO DE LAS CAJAS**

```python
import pandas as pd 
import matplotlib.pyplot as plt
from importlib import reload

import Clases.caja as caja_module
reload(caja_module)
from Clases.caja import Caja

import Clases.producto as producto_module
reload(producto_module)
from Clases.producto import Producto

especificaciones_cajas = pd.read_csv("Datos-finales/especificaciones_cajas.csv")
procurement_cajas = pd.read_csv("Datos-finales/procurement_cajas.csv")
```

Importamos en una lista los tipos de cajas del dataset.

```python
caja_compras_merge = especificaciones_cajas.merge(procurement_cajas,
                                                  on="caja_tipo_id")

cajas = [
    Caja(
        caja_id = row["caja_tipo_id"],
        dim_interior_ancho = row["caja_interior_ancho"],
        dim_interior_largo = row["caja_interior_largo"],
        dim_interior_alto = row["caja_interior_alto"]
    )
    for _, row in caja_compras_merge.iterrows()
]
```

#### **Redimensionamiento con grosor = 5mm**

Para encontrar una cantidad óptima de cajas que entran en un pallet, necesitamos definir un redimensionamiento óptimo para las dimensiones internas de las cajas.

Siguiendo el enunciado, sólo podemos redimensionar un ±10% de las dimensiones internas originales. 

```python
for caja in cajas:
    caja.elegir_grosor(grosor_mm=5)
    
mejor_opcion_por_caja_5mm = {}
combinaciones_ordenadas_por_caja_5mm = {}

for caja in cajas:
    mejor_opcion, combinaciones_ordenadas = caja.buscar_redimensionamiento_optimo()
    mejor_opcion_por_caja_5mm[caja] = mejor_opcion
    combinaciones_ordenadas_por_caja_5mm[caja] = combinaciones_ordenadas
```

Veamos en un dataframe las mejores combinaciones de redimensionamiento por cada tipo de caja.

```python
datos = []

for caja, mejor in mejor_opcion_por_caja_5mm.items():
    datos.append({
        'caja_id': caja.caja_id,
        'alto': mejor[0],
        'ancho': mejor[1],
        'largo': mejor[2],
        'utilizacion_pallet': mejor[3]
    })

df_mejor_opcion_5mm = pd.DataFrame(datos)
df_mejor_opcion_5mm
```

Por otro lado, exportamos en csv todas las combinaciones posibles de redimensionamiento por caja.

```python
data = []

for caja, combinaciones in combinaciones_ordenadas_por_caja_5mm.items():
    for combinacion in combinaciones:
        data.append({
            'caja_id': caja.caja_id,
            'alto': combinacion['alto'],
            'ancho': combinacion['ancho'],
            'largo': combinacion['largo'],
            'utilizacion': combinacion['utilizacion_pallet']
        })

df_todas_combinaciones_5mm = pd.DataFrame(data)
df_todas_combinaciones_5mm.to_csv('Redimensionamientos/redimensionamientos_5mm.csv', index=False)
```
