Análisis exploratorio de los datos

```python
import pandas as pd 
import numpy as np
```

```python
catalogo_productos = pd.read_csv("Datos-originales/catalogo_productos.csv")
especificaciones_cajas = pd.read_csv("Datos-originales/especificaciones_cajas.csv")
operaciones_planta = pd.read_csv("Datos-originales/operaciones_planta.csv")
procurement_cajas = pd.read_csv("Datos-originales/procurement_cajas.csv")
```

```python
catalogo_productos
```

Observamos que faltan algunos valores en relación al tamaño_paquete. Agregamos dichos valores.

```python
cols = (
    catalogo_productos['tamaño_paquete']
    .astype(str)
    .str.upper()
    .str.replace('KG', '', regex=False)
    .str.split('X', expand=True)
)
catalogo_productos['cantidad_paquetes'] = cols[0].str.strip().str.replace(',', '.').astype(float)
catalogo_productos['peso_neto_paquete'] = cols[1].str.strip().str.replace(',', '.').astype(float)
```

```python
especificaciones_cajas
```

Algunas medidas en caja_grosor_mm aparecen con la unidad 'mm', aprovechemos para además transformar la columna en float

```python
especificaciones_cajas['caja_grosor_mm'] = (
    especificaciones_cajas['caja_grosor_mm']
    .astype(str)
    .str.replace('mm', '', regex=False)
    .str.strip()
    .str.replace(',', '.')
    .astype(float)
)
```

Faltan algunos cálculos de la cantidad_cajas_total

```python
especificaciones_cajas['cantidad_cajas_total'] = (
    especificaciones_cajas['cantidad_cajas_alto'] * 
    especificaciones_cajas['cantidad_cajas_largo'] * 
    especificaciones_cajas['cantidad_cajas_ancho']
)
```

```python
procurement_cajas
```

```python
descuento_cols = [col for col in procurement_cajas.columns if 'descuento_planta' in col]

for col in descuento_cols:
    procurement_cajas[col] = (
        procurement_cajas[col]
        .astype(str)
        .str.replace('+', '', regex=False)
        .str.replace('%', '', regex=False)
        .str.strip()
        .str.replace(',', '.')
        .str.replace(r'\s+', '', regex=True)
        .astype(float) / 100
    )
```

```python
# Recalcular costo unitario por planta aplicando el descuento correspondiente
costo_base = pd.to_numeric(procurement_cajas['costo_unitario_base'], errors='coerce').fillna(0)

costo_cols = [c for c in procurement_cajas.columns if c.startswith('costo_unitario_planta_')]

for c in costo_cols:
    plant = c.replace('costo_unitario_planta_', '')
    desc_col = f'descuento_planta_{plant}'
    if desc_col in procurement_cajas.columns:
        descuento = pd.to_numeric(procurement_cajas[desc_col], errors='coerce').fillna(0)
    else:
        descuento = 0
    procurement_cajas[c] = costo_base * (1 + descuento)
```

Exportamos las tablas modificadas en la carpeta "Datos-limpiados"

```python
catalogo_productos.to_csv("Datos-limpiados/catalogo_productos.csv", index=False)
especificaciones_cajas.to_csv("Datos-limpiados/especificaciones_cajas.csv", index=False)
operaciones_planta.to_csv("Datos-limpiados/operaciones_planta.csv", index=False)
procurement_cajas.to_csv("Datos-limpiados/procurement_cajas.csv", index=False)
```
