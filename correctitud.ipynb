```python
def comparar_csv_por_columna(df1, df2, columna):

    df1 = df1.reset_index(drop=True)
    df2 = df2.reset_index(drop=True)

    resultado = pd.DataFrame({
        "fila": range(1, max(len(df1), len(df2)) + 1),
        f"{columna}_archivo1": df1[columna].fillna(""),
        f"{columna}_archivo2": df2[columna].fillna(""),
    })

    resultado["coincide"] = resultado[f"{columna}_archivo1"] == resultado[f"{columna}_archivo2"]
    return resultado
```

```python
df_comparacion = comparar_csv_por_columna(procurement_cajas, procurement_cajas2, "costo_total_base")

df_comparacion
```
